```python
# Creating XML files for SAVI 2018 Data
import csv

csvFile = 'CHR2019.csv'
xmlFile = 'CHR2019.xml'

csvData = csv.reader(open(csvFile))
xmlData = open(xmlFile, 'w')
xmlData.write('<?xml version="1.0" encoding="UTF-8"?>' + "\n")
# there must be only one top-level tag
xmlData.write('<chr_data>' + "\n")

rowNum = 0
for row in csvData:
    if rowNum == 0:
        tags = row
    else: 
        xmlData.write('<dataValueSet' + ' xmlns="http://dhis2.org/schema/dxf/2.0" dataSet="KigDfFRHq4Q"' + ' completeDate="2021-07-11" period="2019" orgUnit="' + row[0] +'">' + "\n")
        for i in range(1, len(tags)):
            xmlData.write('    ' + '<dataValue dataElement="' + tags[i] + '"' + ' value="'+ row[i] + '"' + '/>' + "\n")
        xmlData.write('</dataValueSet>' + "\n")
    rowNum +=1

xmlData.write('</chr_data>' + "\n")
xmlData.close() 
```
