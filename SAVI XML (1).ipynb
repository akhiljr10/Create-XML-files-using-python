```python
import pandas as pd
```

```python
df = pd.read_csv('savixml.csv')
```

```python
dropcols = ['GEOGRAPHY_TYPE', 'GEOLOCID']
```

```python
df1 = df.drop(columns = dropcols)
```

```python
df2018 = df1[df1['DATAYEAR'] == 2018]
```

```python
df2019 = df1[df1['DATAYEAR'] == 2019]
```

```python
df2018 = df2018.drop(columns = 'DATAYEAR')
df2019 = df2019.drop(columns = 'DATAYEAR')
```

```python
df2018 = df2018.round(3)
```

```python
df2019 = df2019.round(3)
```

```python
df2018.to_csv('savi2018.csv', encoding='utf-8', index=False)
```

```python
df2019.to_csv('savi2019.csv', encoding='utf-8', index=False)
```

```python
# Creating XML files for SAVI 2018 Data
import csv

csvFile = 'savi2018.csv'
xmlFile = 'savi2018.xml'

csvData = csv.reader(open(csvFile))
xmlData = open(xmlFile, 'w')
xmlData.write('<?xml version="1.0" encoding="UTF-8"?>' + "\n")
# there must be only one top-level tag
xmlData.write('<savi_data>' + "\n")

rowNum = 0
for row in csvData:
    if rowNum == 0:
        tags = row
    else: 
        xmlData.write('<dataValueSet' + ' xmlns="http://dhis2.org/schema/dxf/2.0" dataSet="gYcWLyWOqeT"' + ' completeDate="2021-12-10" period="2018" orgUnit="' + row[0] +'">' + "\n")
        for i in range(1, len(tags)):
            xmlData.write('    ' + '<dataValue dataElement="' + tags[i] + '"' + ' value="'+ row[i] + '"' + '/>' + "\n")
        xmlData.write('</dataValueSet>' + "\n")
    rowNum +=1

xmlData.write('</savi_data>' + "\n")
xmlData.close() 
```
